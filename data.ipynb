```python
import numpy as np
import matplotlib.pyplot as plt
  
  
# Text file data converted to integer data type
File_data = np.loadtxt("data/PositionControl_4.txt", dtype=float)

time = []
distance = []
orientation_error = []
count = 0
for data in File_data:
    if count % 3 == 0:
        time.append(File_data[count])
    if count % 3 == 1:
        distance.append(File_data[count])
    if count % 3 == 2:
        orientation_error.append(File_data[count])
    count += 1
```

```python
# plot the curve using Matplotlib
plt.plot(time, distance)

# set the x and y axis labels
plt.xlabel('Time')
plt.ylabel('Distance')

# set the title of the plot
plt.title('Time Vs. Distance')
# show the plot
plt.show()

print(distance)
```

```python
# plot the curve using Matplotlib
plt.plot(time, orientation_error)

# set the x and y axis labels
plt.xlabel('Time')
plt.ylabel('Orientation Error')

# set the title of the plot
plt.title('Time Vs. Orientation Error')

# show the plot
plt.show()
```
