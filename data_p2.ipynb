```python
import numpy as np
import matplotlib.pyplot as plt
  
  
# Text file data converted to integer data type
File_data = np.loadtxt("data/TrajectoryTrackingRectangular_2.txt", dtype=float)

time = []
robot_x = []
robot_y = []
target_x = []
target_y = []
count = 0
for data in File_data:
    if count % 5 == 0:
        time.append(File_data[count])
    if count % 5 == 1:
        robot_x.append(File_data[count])
    if count % 5 == 2:
        robot_y.append(File_data[count])
    if count % 5 == 3:
        target_x.append(File_data[count])
    if count % 5 == 4:
        target_y.append(File_data[count])
    count += 1
```

```python
# plot the curve using Matplotlib
plt.plot(time, robot_x, label = 'robot_x')
plt.plot(time, target_x, label = 'desired_x')

# set the x and y axis labels
plt.xlabel('Time interval')
plt.ylabel('X coordinate')
# add a legend
plt.legend()
# set the title of the plot
plt.title('Time Vs. X diff')
# show the plot
plt.show()
```

```python
# plot the curve using Matplotlib
plt.plot(time, robot_y, label = 'robot_y')
plt.plot(time, target_y, label = 'desired_y')

# set the x and y axis labels
plt.xlabel('Time interval')
plt.ylabel('Y coordinate')
# add a legend
plt.legend()
# set the title of the plot
plt.title('Time Vs. Y diff')
# show the plot
plt.show()
```
